#### 2. Check and Drop missing or null values

```python
print(cars.isnull().sum())
```

```python
cars = cars.dropna() 
cars.count()
```

#### 3. What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
avg_msrp = cars.groupby('Make')['MSRP'].mean().round()
top_10_avg_msrp = avg_msrp.sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_avg_msrp.values, y=top_10_avg_msrp.index, palette="viridis")

plt.title("Top 10 Average MSRP per Car Brand")
plt.xlabel("Average MSRP")
plt.ylabel("Car Brand")
plt.tight_layout()
plt.show()
```

#### 4. How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
avg_msrp_per_year = cars.groupby('Year')['MSRP'].mean().round()

plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_msrp_per_year.index, y=avg_msrp_per_year.values, marker='o', linewidth=2)

plt.title("Average MSRP Over the Years")
plt.xlabel("Year")
plt.ylabel("Average MSRP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 5. How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
cars['Average_MPG'] = (cars['city mpg'] + cars['highway MPG']) / 2

avg_stats_per_year = cars.groupby('Year')[['Engine HP', 'Average_MPG']].mean().round()

plt.figure(figsize=(12, 6))

sns.lineplot(x=avg_stats_per_year.index, y=avg_stats_per_year['Engine HP'], label='Engine HP', marker='o')

sns.lineplot(x=avg_stats_per_year.index, y=avg_stats_per_year['Average_MPG'], label='Average MPG', marker='o')

plt.title("Trends in Engine HP and MPG Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### 7. Which vehicle styles are sold the most? show with vertical bar plot

```python
# Plot
plt.figure(figsize=(12,6))
style_counts.plot(kind='bar', color='mediumslateblue')

# Formatting
plt.title('Most Sold Vehicle Styles', fontsize=14)
plt.xlabel('Vehicle Style')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 8. Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
filtered_cars = cars[(cars['MSRP'] >= 40000) & (cars['Popularity'] >= 1000)]

brand_counts = filtered_cars['Make'].value_counts()
brand_counts

# Plot
plt.figure(figsize=(12,6))
brand_counts.plot(kind='bar', color='darkcyan')

# Formatting
plt.title('Car Brands with Vehicles ≥ $40,000 and Popularity ≥ 1000')
plt.xlabel('Car Brand')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
